```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from pylab import *
figure()
x = np.linspace(0, 5, 10)
y = x ** 2
plot(x, y, color='gray', marker='*', linestyle='dashed', linewidth=1, markersize=6)
xlabel('x')
ylabel('y')
title('test')
show()
```

```python
# 子图（行，列，位置）
subplot(2,1,1)
plot(x, y, 'r--')
xlabel('x')
ylabel('y')
title('test1')
subplot(2,1,2)
plot(y, x, 'g*-')
xlabel('age')
ylabel('height')
title('boy')
show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
axes.plot(x, y, 'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes
axes3 = fig.add_axes([0.4, 0.7, 0.])

# main figure
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('title')

# insert
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('insert title')
```
